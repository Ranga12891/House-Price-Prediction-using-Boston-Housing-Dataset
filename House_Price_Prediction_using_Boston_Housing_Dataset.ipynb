**Detecting Outliers**

```python
# Dropping ID Column as it is a unique Column
Train_DS.drop('ID',axis = 1,inplace = True)
Train_DS.head(5)
```

```python
Columns = Train_DS.columns.values
Outliers = []
Columns
```

```python
for i in Columns:
  Q1 = Train_DS[i].quantile(.75)
  Q3 = Train_DS[i].quantile(.25)
  IQR = Q1 - Q3
  Rng1 = Q3 - (IQR * 1.5)
  Rng2 = Q3 + (IQR * 1.5)
  Outlier = True if Train_DS[i].min() < Rng1 or Train_DS[i].max() > Rng2 else False
  Outliers.append(Outlier)
  # Correcting the Outliers
  Train_DS[i].loc[(Train_DS[i] < Rng1) ] = Rng1
  Train_DS[i].loc[(Train_DS[i] > Rng2) ] = Rng2
```

```python
Outliers.count(True)
```

```python
len(Columns)
```

There were 11 columns with Outliers out of 14. Those were corrected

```python
Train_DS.describe()
```

Data CLeaning is done.
There is no need to encode the data.
Now we can Split & Scale the data

```python
X = Train_DS.drop('MEDV',axis = 1)
```

```python
Y = Train_DS.MEDV
```

```python
# Spltting
from sklearn.model_selection import train_test_split #import the required function
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3,random_state=1)

# Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

This is Supervised Learning Problem. Since Target is Continuos this is Regression Problem.

# **We will Apply Linear, Lasso & Ridge Regression Model.**

**Linear Regression**

```python
from sklearn.linear_model import LinearRegression #importing all the required functions
regressor = LinearRegression() # spredicted score = m * hours + c  
regressor.fit(X_train_scaled, y_train) #Learning happens - GD is done and we get the final values of m and c
y_pred = regressor.predict(X_test_scaled)
y_pred.shape
```

```python
LR_R2_Score = round(regressor.score(X_test_scaled,y_test),3)
LR_R2_Score
```

**Lasso Regression**

```python
from sklearn.linear_model import Lasso
lr = Lasso(alpha = 0.1) # Hyper paramerer Lambda
lr.fit(X_train_scaled, y_train)
y_pred = lr.predict(X_test_scaled)
```

```python
Lasso_R2_Score = round(lr.score(X_test_scaled, y_test),3)
Lasso_R2_Score
```

**Ridge Regression**

```python
from sklearn.linear_model import Ridge
Ridge_LR = Ridge(alpha = 1) # Hyper paramerer Lambda
Ridge_LR.fit(X_train_scaled, y_train)
y_pred = Ridge_LR.predict(X_test_scaled)
```

```python
Ridge_R2_Score = round(Ridge_LR.score(X_test_scaled, y_test),3)
Ridge_R2_Score
```

Out of All Modles Ridge is having better Score.

So Selecting Ridge Regression as final model
